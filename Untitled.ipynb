```python
def my_pizza_generator():
    yield "Pepperoni"
    yield "Mushroom"
    yield "Extra Cheese"

# This doesn't make the pizzas yet; it just creates the "chef"
chef = my_pizza_generator()

print(next(chef)) # Output: Pepperoni (The chef makes one and pauses!)
print(next(chef)) # Output: Mushroom (The chef resumes, makes one, and pauses again!)
```
